```python
with open("body.html", "r") as body:
    content = body.read()
    print(content)
    
soup  = BeautifulSoup(content,'lxml')
```

```python
title_element = soup.find("span", {"id": "productTitle"})

# Extract the product title text
product_title = title_element.get_text(strip=True)
```
